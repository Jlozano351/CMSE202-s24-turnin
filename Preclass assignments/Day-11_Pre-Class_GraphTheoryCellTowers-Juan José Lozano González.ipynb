## Defining the problem

The signals from different cellphone towers that are close together can interfere.

## Step 1: Generate random cell towers. 

To start out, we're going to keep things simple and use artificial data to create our cell phone tower network. Specifically, we'll just populate a network of towers randomly.

**&#9989;  Do This:** **Write a function** to randomly create a 2D numpy array with $N$ rows representing cell towers and 2 columns representing the $x$ (longitude) and $y$ (latitude) positions of the cell towers on a map. You may assume that the map is a 2D plane and not a sphere. You can also assume that the "latitude" and "longitude" values are all in the range of 0 to 1.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
class Locations:
    def __init__(self, number_of_towers):
        """
        Initialize the Locations class.

        Parameters:
        - number_of_towers (int): Number of towers for which locations need to be generated.
        """
        self.number_of_towers = number_of_towers
        self._locations = self._generate_points()  # Generate points upon initialization

    def _generate_points(self):
        """
        Generate random locations for the towers.

        Returns:
        - locations (numpy.ndarray): 2D array of shape (2, number_of_towers) representing tower locations.
        """
        locations = np.zeros([2, self.number_of_towers])  # Create an array to store locations
        locations[0, :] = np.random.rand(self.number_of_towers)  # Generate random x-coordinates
        locations[1, :] = np.random.rand(self.number_of_towers)  # Generate random y-coordinates
        return locations

    @property
    def locations(self):
        """
        Property to access the generated tower locations.

        Returns:
        - _locations (numpy.ndarray): 2D array of shape (2, number_of_towers) representing tower locations.
        """
        return self._locations
    
    def plot(self):
        """
        Plot the tower locations.
        """
        plt.scatter(self.locations[0, :], self.locations[1, :])  # Scatter plot of tower locations
        plt.xlabel('longitude')  # Label for x-axis
        plt.ylabel('latitude')  # Label for y-axis
        plt.title('Randomly generated cell towers')  # Title for the plot

    def create_nearest_neighbor_list(self):
        """
        creates a nearest_neighbor_list
        """

        def distance(p1,p2):
            '''Returns the distance betwen two numpy points'''
            return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

        def find_nearest_point(index):
            best_distance = np.inf
            index_of_shortest_distance = None
            for i in range(len(self.locations[0])):
                if i == index:
                    continue
                else:
                    distance_value = distance(self.locations[:,index],self.locations[:,i])
                    if distance_value < best_distance:
                        best_distance = distance_value
                        index_of_shortest_distance = i

            return index_of_shortest_distance
    
        neighbor_list = []
        for i in range(len(self.locations[0])):
            neighbor_list.append(find_nearest_point(i))

        return neighbor_list

    def plot_nearest_neighbor(self):
        """
        Plot the nearest neighbor graph for the location of the tower cells.
        """
        plt.scatter(self.locations[0, :], self.locations[1, :])  # Scatter plot of tower locations
        plt.xlabel('longitude')  # Label for x-axis
        plt.ylabel('latitude')  # Label for y-axis
        plt.title('nearest neighbor graph')  # Title for the plot

        # Create the neighbor list:
        neighbor_list = self.create_nearest_neighbor_list()

        # Plots lines betweeen nearest neighbors
        for i in range(len(self.locations[0])):
            point = self.locations[:, i]
            nearest_index = neighbor_list[i]
            nearest_neighbor = self.locations[:, nearest_index]
            array = np.vstack((point, nearest_neighbor))  
            plt.plot(array[:, 0], array[:, 1], color = "Blue")  
````
